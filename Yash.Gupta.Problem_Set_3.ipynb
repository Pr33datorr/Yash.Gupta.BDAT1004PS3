Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

Step 2. Import the dataset from this address.

Step 3. Assign it to a variable called users

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')
print(users)
```

Step 4. Discover what is the mean age per occupation

```python
#dataframe is created for the data imported using pandas
df = pd.DataFrame(users)
#Find the mean age and group by occupation.
mean_age = df.groupby('occupation')['age'].mean()
print(mean_age)
```

Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
#To find the number of male and group by occupation
male_ratio= df[df['gender']=='M'].groupby('occupation')['gender'].count()
```

```python
#To find the ratio with by dividing the count of men by total population
ratio = male_ratio / df.groupby('occupation')['gender'].count()
#sort the values in descending
sorted_values= ratio.sort_values(ascending=False)
#Print the values
print (sorted_values)
```

Step 6. For each occupation, calculate the minimum and maximum ages

```python
#To find the min and max using aggregate function
maxmin_age = df.groupby('occupation').agg({'age':['min', 'max']})
#Print the result
print(maxmin_age)
```

Step 7. For each combination of occupation and sex, calculate the mean age

```python
#To find the mean age for male and female corresponding to each occupation
mean_age = df.groupby(['occupation','gender'])['age'].mean()

#Print the results
print(mean_age)
```

Step 8. For each occupation present the percentage of women and men

```python
# To calculate the count of male and female with occupation
gender_count = df.groupby(['occupation', 'gender'])['gender'].count()

# TO calculate the count corresponding to each occupation
count_occupation = df.groupby('occupation')['gender'].count()

# To calculate the percentage male and female based on occupation
percent = (gender_count / count_occupation) * 100

#Print the result
print(percent)
```
